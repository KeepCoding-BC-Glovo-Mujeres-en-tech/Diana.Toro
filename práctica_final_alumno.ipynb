```python
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

clave = list(articulos.keys())
valor = list(articulos.values())

indice = [clave.index(x) for x in clave]
```

```python
# A. Una función que calcule el precio total de la compra
def precio(compra, clave, valor):
  i = 0
  suma = 0
  for x in compra:
    if x in clave:
      suma += valor[clave.index(x)]
    i +=1
  return suma
    
precio_total = precio(compra, clave, valor)
print(f'El Precio total de la compra es : {precio_total}')
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
def descuento(precio):
  valor = (precio - (precio * 0.10))
  return valor

valor_descuento = descuento(precio_total)
print(f'El Precio total de la compra con 10% de descuento es : {valor_descuento}')
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def compra_mayor_100(precio):
  if precio > 100:
    valor = (precio - (precio * 0.10))
  else:
    valor = precio
  return valor

valor_descuento = compra_mayor_100(precio_total)
print(f'El Precio total de la compra con 10% de descuento es : {valor_descuento}')
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
def impuesto_iva(precio):
  valor = precio + (precio * 0.21)
  return valor

valor_impuesto = impuesto_iva(precio_total)
print(f'El Precio total de la compra con el iva del 21% es : {valor_impuesto}')
```

```python
# E.lista los artículos cuyo precio es superior a 20€
def precio_mayor_20(articulos):
  new_list = []
  for y in valor:
    if y > 20:
      new_list.append(clave[valor.index(y)])
  return new_list

lista = precio_mayor_20(articulos)

print(f'Los árticulos que superan 20€ son:')
print(lista)
```

```python
'''
14. Dada una lista de tuplas con el nombre de un alumno, apellidos, curso y sus notas. 

 Define una funcion que reciba el curso y saque una lista en la que aparezca el nombre y apellidos y el promedio de sus notas.
 Puedes definir más funciones si lo consideras necesario.
'''

alumnos = [('Juan', 'Pérez', '1', [5, 6, 7, 8, 9]),
            ('Ana', 'García', '2', [5, 6, 7, 8, 9]),
            ('Luis', 'González', '1', [5, 6, 7, 8, 9]),
            ('María', 'Martínez', '2', [5, 6, 7, 8, 9]),
            ('Pedro', 'Rodríguez', '1', [5, 6, 7, 8, 9]),
            ('Lucía', 'Hernández', '2', [5, 6, 7, 8, 9]),
            ('Marta', 'López', '1', [5, 6, 7, 8, 9]),
            ('Sara', 'Díaz', '2', [5, 6, 7, 8, 9]),
            ('Javier', 'Sánchez', '1', [5, 6, 7, 8, 9]),
            ('Miguel', 'Fernández', '2', [5, 6, 7, 8, 9]),
            ('Sergio', 'Jiménez', '1', [5, 6, 7, 8, 9]),
            ('Sandra', 'Ruiz', '2', [5, 6, 7, 8, 9]),
            ('Pablo', 'Álvarez', '1', [5, 6, 7, 8, 9]),
            ('María', 'Gómez', '2', [5, 6, 7, 8, 9]),
]
```

```python
from functools import reduce
def estudio(alumnos):

  nombres = []
  apellidos = []
  curso = []
  notas = []

  i = 0
  while i < len(alumnos):
    nombres.append(alumnos[i][0])
    apellidos.append(alumnos[i][1])
    curso.append(alumnos[i][2])
    notas.append(alumnos[i][3])
    i += 1
  
  promedio = []
  for x in notas:
    promedio.append(reduce(lambda a,b: a+b, x, 0)/5)

  dato = ''
  while not dato.isnumeric() or int(dato) not in range(1,3):
    dato = input('Ingrese el curso: ')
  
  n_l = []
  for i in range(len(alumnos)):
    if curso[i] == dato:
      n_l.append([nombres[i],apellidos[i],curso[i],promedio[i]])
    else:
      continue
    
  return n_l

resultado = estudio(alumnos)
print('NOMBRE', 'APELLIDO', 'CURSO', 'PROMEDIO')
for i in range(len(resultado)):
  print(resultado[i], end=' ')
  print()
```
